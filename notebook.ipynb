```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras import datasets
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
import numpy as np
```

# Load and Preprocess the Dataset

```python
(train_images, train_labels), (test_images, test_labels) = datasets.fashion_mnist.load_data()
```

```python
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
# Reshape the images to include the channel dimension for grayscale (28x28x1)
train_images = np.expand_dims(train_images, axis=-1)  # (60000, 28, 28, 1)
test_images = np.expand_dims(test_images, axis=-1)    # (10000, 28, 28, 1)
```

```python
# Convert grayscale images to RGB by repeating the single channel 3 times
train_images = np.repeat(train_images, 3, axis=-1)    # (60000, 28, 28, 3)
test_images = np.repeat(test_images, 3, axis=-1)      # (10000, 28, 28, 3)
```

```python
# Convert labels to categorical (one-hot encoding)
train_labels_cat = to_categorical(train_labels, 10)
test_labels_cat = to_categorical(test_labels, 10)

# Define a function to resize images
def preprocess_image(image, label):
    image = tf.image.resize(image, (224, 224))  # Resize to MobileNetV2 input size
    return image, label

# Convert the data into tf.data.Dataset for better performance
train_dataset = tf.data.Dataset.from_tensor_slices((train_images, train_labels_cat))
test_dataset = tf.data.Dataset.from_tensor_slices((test_images, test_labels_cat))

# Apply the preprocessing function (resize) to the dataset
train_dataset = train_dataset.map(preprocess_image).batch(32).prefetch(tf.data.AUTOTUNE)
test_dataset = test_dataset.map(preprocess_image).batch(32).prefetch(tf.data.AUTOTUNE)
```

```python
print("Train images shape:", train_images.shape)
```

```python
print(f'Test images shape: {test_images.shape}')
```

# MobileNetV2 model with pre-trained weights from ImageNet

```python
# Load the MobileNetV2 model with pre-trained weights from ImageNet
base_model = MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
```

```python
# Freeze the base model layers
base_model.trainable = False
```

```python
# Build the model
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')  # 10 classes for Fashion MNIST
])
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Summary of the model
model.summary()
```

```python
# Train the model using the datasets
history = model.fit(train_dataset, 
                    epochs=10, 
                    validation_data=test_dataset)
```

# Evaluation, Prediction, and Visualization

```python
import matplotlib.pyplot as plt
```

```python
# 1. Evaluate the model
test_loss, test_acc = model.evaluate(test_dataset)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_acc}")
```

```python
# 2. Make predictions on the test dataset
# Since the dataset is batched, we need to unbatch it for predictions
unbatched_test_images = []
unbatched_test_labels = []
for images, labels in test_dataset.unbatch().take(100):  # Take 100 examples for visualization
    unbatched_test_images.append(images)
    unbatched_test_labels.append(labels)
```

```python
# Convert list to numpy array for easier handling
unbatched_test_images = np.array(unbatched_test_images)
unbatched_test_labels = np.array(unbatched_test_labels)

# Make predictions
predictions = model.predict(unbatched_test_images)
predicted_labels = np.argmax(predictions, axis=1)
true_labels = np.argmax(unbatched_test_labels, axis=1)
```

```python
# 3. Visualize the predictions
def plot_predictions(images, true_labels, predicted_labels, class_names, num=10):
    """Visualize some of the test images with their true and predicted labels"""
    plt.figure(figsize=(15, 15))
    for i in range(num):
        plt.subplot(5, 5, i+1)
        
        # Rescale the images to 0-255 if needed and cast to uint8
        img = np.clip(images[i] * 255.0, 0, 255).astype('uint8')
        
        plt.imshow(img)
        plt.title(f"True: {class_names[true_labels[i]]}\nPred: {class_names[predicted_labels[i]]}")
        plt.axis('off')
    plt.show()

# Plot a few test images along with their true and predicted labels
plot_predictions(unbatched_test_images, true_labels, predicted_labels, class_names)
```
